```python
import pandas as pd

df = pd.read_csv('data/sales.csv', sep=';', decimal=',')
df.head(5)
```

```python
df.info()
```

```python
# Converter data
df['Date'] = pd.to_datetime(df['Date'], yearfirst=True, errors='coerce')

df.dropna()
# df['Total'] = df['Total'].fillna(0)
df.info()
```

## Análises temporais

```python
df['City'] = df['City'].str.title().str.strip()
df['Customer type'] = df['Customer type'].str.capitalize().str.strip()
df['Payment'] = df['Payment'].str.capitalize().str.strip()

df.info()
```

```python
df['Month'] = df['Date'].dt.to_period('M').astype(str)
# df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
df['Day'] = df['Date'].dt.day
df['Weekday'] = df['Date'].dt.day_name()


# Ticket Médio (Total / Quantity)
df['Ticket_Medio'] = df['Total'] / df['Quantity']

df
```

```python
df.info()
```

```python
import sqlite3

con = sqlite3.connect('supermarket.db')

df.to_sql('SalesSuperMarket', con=con, if_exists='replace', index=False)

with open('saida.sql', 'w', encoding='utf-8') as f:
    for linha in con.iterdump():
        f.write(f'{linha}\n')
        
con.close()
```
