```python
import pandas as pd
```

```python
df_sales = pd.read_csv('data/sales.csv', sep=';', decimal=',', index_col=0)

df_sales.head()
```

```python
# df_sales.info()
df_sales.head()
# df_sales.describe(include='all')  # Inclui colunas categóricas também
# df_sales.isnull().sum()  # Checar valores nulos
```

```python
df_sales['Date'] = pd.to_datetime(df_sales['Date'])

df_sales.info()
```

```python
df_sales = df_sales.sort_values('Date')
df_sales
```

```python
df_good_rate = df_sales[(df_sales['Rating'] >= 7) & (df_sales['Quantity'] >= 7)]
df_good_rate
```

```python
df_good_rate['Customer type'].value_counts(normalize=True)
```

```python
df_sales.groupby('Date')['Total'].sum().sort_values(ascending=False).head(10)
```

```python
df_sales.groupby('Product line')['Rating'].mean().sort_values(ascending=False)
```

```python
df_sales['Hour'] = pd.to_datetime(df_sales['Time']).dt.hour
df_sales['Hour'].value_counts().sort_index().plot(kind='bar', title='Vendas por Hora do Dia')
```

```python
df_sales.groupby('Product line')['Rating'].mean().sort_values(ascending=False)
```

```python
import sqlite3

con = sqlite3.connect('supermarket.db')

df_good_rate.to_sql('minhatabela', con=con, if_exists='replace', index=False)

with open('saida.sql', 'w', encoding='utf-8') as f:
    for linha in con.iterdump():
        f.write(f'{linha}\n')
        
con.close()

```

```python
df_good_rate.value_counts('Gender')
df_good_rate.value_counts('City')
df_good_rate.value_counts('Branch', 'City')
df_good_rate.value_counts('Customer type', 'Gender')

```

```python
df_sales.groupby('City')['Total'].sum().sort_values(ascending=False)
```

```python
df_sales['Month'] = df_sales['Date'].dt.to_period('M')
receita_mes = df_sales.groupby('Month')['Total'].sum()

receita_mes.plot(kind='line', title='Evolução das Vendas por Mês')

```
